```python
import streamlit as st
from transformers import pipeline

# Initialize the chatbot model
chatbot = pipeline("conversational", model="facebook/blenderbot-400M-distill")

st.title("Chatbot using Streamlit")

# Placeholder for chat history
chat_history = []

def chat_with_bot(user_input, chat_history):
    # Append user input to chat history
    chat_history.append(user_input)

    # Create a conversation object with the chat history
    conversation = chatbot.conversation(chat_history)

    # Get the bot response
    bot_response = chatbot(conversation)

    # Append bot response to chat history
    chat_history.append(str(bot_response))

    return str(bot_response), chat_history

user_input = st.text_input("You: ", "")

if st.button("Send"):
    if user_input:
        response, chat_history = chat_with_bot(user_input, chat_history)
        st.write(f"Bot: {response}")

# Display the chat history
if chat_history:
    st.write("### Chat History")
    for i, message in enumerate(chat_history):
        if i % 2 == 0:
            st.write(f"You: {message}")
        else:
            st.write(f"Bot: {message}")
```

```python
# Import necessary libraries
from transformers import pipeline, Conversation

# Initialize the chatbot model
chatbot = pipeline("conversational", model="facebook/blenderbot-400M-distill")

# Placeholder for chat history
chat_history = []

def chat_with_bot(user_input, chat_history):
    # Append user input to chat history
    chat_history.append(user_input)

    # Create a conversation object with the chat history
    conversation = Conversation(user_input)

    # Get the bot response
    bot_response = chatbot(conversation)

    # Extract the response text
    response_text = bot_response.generated_responses[-1]

    # Append bot response to chat history
    chat_history.append(response_text)

    return response_text, chat_history

# Simulate interaction in Colab
while True:
    # Get user input
    user_input = input("You: ")

    # Exit loop if user types 'exit'
    if user_input.lower() == 'exit':
        print("Exiting chat...")
        break

    # Get response from chatbot
    response, chat_history = chat_with_bot(user_input, chat_history)
    print(f"Bot: {response}")

    # Display the chat history
    print("\n### Chat History ###")
    for i, message in enumerate(chat_history):
        if i % 2 == 0:
            print(f"You: {message}")
        else:
            print(f"Bot: {message}")
    print("####################\n")
```
